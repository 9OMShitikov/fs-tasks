## Сравнение времён работы программы при разных параметрах

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv('./results.csv', sep=',')
```

Разделим результаты на две группы:

```python
df_small_size = df[df["size_val"] <= 1024 * 1024]
df_huge_size = df[df["size_val"] >= 1024 * 1024]
```

 Создадим функцию для построения графиков:

```python
def plot_times(df):
    plt.figure(figsize=(12, 8))

    for name in df.columns[2:]:
        plt.plot(
            df["size_val"],
            df[name],
            label=name
        )

    plt.rc('ytick', labelsize=13)
    plt.ylabel('Время работы, сек', fontsize=15)
    plt.xlabel('Размер копируемого файла', fontsize=15)
    plt.xticks(df["size_val"], df["size_name"], fontsize=13)
    plt.legend(loc='best', fontsize=13)
    plt.tight_layout()
    plt.grid(ls=':')
    plt.title("Время работы в зависимости от размера копируемого файла", fontsize=15)
    plt.show()
    
```

Проверим работу программы при размере копируемого файла меньше 1MiB. Команда cp консоли добавлена для сравнения.

```python
plot_times(df_small_size)
```

Теперь проверим работу программы при размере копируемого файла больше 1MiB.

```python
plot_times(df_huge_size)
```

 Мы видим, что чтение 4 запросами по 256K - наилучший вариант.
